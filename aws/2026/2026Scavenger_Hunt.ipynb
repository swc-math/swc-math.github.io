## A very Magma scavenger hunt



## 1

Let $E$ be the elliptic curve 

$$
E \colon y^2 = x^3-x-2
$$

Let $n_1$ be the rank of $E$. 



## 2

Let $n_2$ be the largest prime factor of the automorphism group of the $E_8$ lattice. 


## 3

Let $G = \mathrm{SL}_2(\mathbb{F}_5)$, and define the subgroup $H \leq G$ by 

$$
H = \bigg \langle \begin{bmatrix} 0 & 4 \\ 1 & 0 \end{bmatrix},   \begin{bmatrix} 2 & 0 \\ 0 & 3 \end{bmatrix}, \begin{bmatrix} 4 & 0 \\ 0 & 4 \end{bmatrix} \bigg\rangle.
$$

Let $n_3 = [G \colon H]$. 



## 4

Let $w \in \mathbb{Z}_{19}$ \(the 19\-adic numbers\) satisfy 

\(1\) $w^2 = 5$,

\(2\) $w \equiv 9 \left( \mathrm{mod} ~ ~  19 \right)$.

Let $n_4$ be the coefficient of $19^{10}$ in the $19$\-adic expansion of $w$. 



## 5

Let $A$ be the quaternion algebra $\left( \frac{-2,-5}{\mathbb{Q}} \right)$. Let $n_{5}$ be the order of the unit group of a maximal order of $A$.


## 6

Define the recurrence $a_n$ by $a_1 = 1$, $a_2 = 3$, and $a_{n+2} = na_{n+1} - a_{n}$. Let $n_6$ be $a_{100} \left( \mathrm{mod}~ ~  23 \right)$. 


## 7

Let $G = \mathrm{GL}_3(\mathbb{F}_5)$ act on $\mathbb{F}_5^3$ in the usual way. Let $H \leq G$ be the subgroup stabilising the line generated by $(1, 0, 1)$. Let $n_7$ be the largest prime factor of $|H|$. 


<span style='font-size:x-large'>8</span>

Let $n_8$ be the number of Carmichael numbers less than $60,000$. 


<span style='font-size:x-large'>9</span>

Let $\tau = \sqrt{-57}$. The value $j(\tau)$ is algebraic, where $j$ is the $j$\-invariant function on the upper half\-plane. Write $f(z)$ for the minimal polynomial of $j(\tau)$ over the integers. Let $n_9$ be the number of prime factors of the constant term of $f(z)$. 


<span style='font-size:x-large'>10</span>

Let $d = n_6 - n_3$, and let $K = \mathbb{Q}(\sqrt{d})$. Write $w = \frac{1+\sqrt{d}}{2}$. Define the elliptic curve over $K$: 


$$
E \colon {y}^2+\left(w+n_1\right){x}{y}+{y}={x}^{3}-\tfrac{n_5 w}{n_9}{x}^{2}+\left(\tfrac{n_6}{3}w - n_7 \right){x}-n_5.
$$

Let $n_{10} = \# E(K)$ . 



<span style='font-size:x-large'>11</span> 

Let $m$ be the smallest integer such that $\phi(\phi(\phi(m))) = 8$, where $\phi$ is the Euler phi \(or totient\) function. Let $n_5$ be the integer such that $\mathrm{prime}(n_{11}) = m$, where $\mathrm{prime}(r)$ gives the $r$th prime number by size. 


<span style='font-size:x-large'>12</span>

Write $\mu(n)$ for the Möbius function. Let $n_{12} = \prod_{k=1}^{11} 2^{\mu(n_k)}$. 


Each of the $n_i$ corresponds to a letter of the English alphabet via the usual association $1 \leftrightarrow a$, $2 \leftrightarrow b$, ..., $26 \leftrightarrow z$. Using this map, unscramble the letters to make the _secret phrase_.
