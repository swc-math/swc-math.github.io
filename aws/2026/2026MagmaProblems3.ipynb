## Question 4: Bianchi modular forms \(hard extension problem\)

Bianchi modular forms are modular forms for $\mathrm{PSL}_2(\mathbb{Z}_F)$, where $F$ is an imaginary quadratic field. Modular symbols algorithms have been developed for these groups, although the added complications of working over a number field mean the general algorithms are much more involved in this setting. Over $\mathbb{Q}(\sqrt{-3})$ the situation is relatively simple, since the field is Euclidean \(and so the class number is 1, and we still have a simple continued fraction algorithm\). Just as classical modular forms are functions on the upper half\-plane $\mathbf{H}^2$, Bianchi modular forms are functions on the three\-dimensional upper half\-_space_ $\mathbf{H}^3 \simeq \mathbb{C} \times \mathbb{R}_{>0}$.
