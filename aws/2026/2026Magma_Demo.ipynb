<span style='font-size:x-large'>A brief introduction to Magma</span>

<span style='font-size:small'>Magma is a </span><span style='font-size:small'>**computational algebra system**</span><span style='font-size:small'> developed at the University of Sydney. It is designed to be an environment where mathematically rigorous computations can be done quickly and easily. It can be used to work with groups, rings, fields, modules, algebras, schemes, curves, graphs, and more. </span>

<span style='font-size:small'>Objects like integers, polynomials and orders, as well as associated rings and fields, are very easy to define. </span>



```magma
Z := Integers();
n := 1;
n in Z;
n + n eq 2;

P<x> := PolynomialRing(Z);
f := x^3 + 2*x^2 - x - 1;
g := P![ 1, 2, -3, -6, 2, 4, 1 ];
f^2 eq g;

F := NumberField(f);
IsIsomorphic(GaloisGroup(F),CyclicGroup(3));
```

We can also define objects from geometry, such as elliptic and hyperelliptic curves. Magma allows you to compute many of the important invariants associated to these objects. 


```magma
E := EllipticCurve([4,-3]);
C := HyperellipticCurve(x^6-3*x+1);

E;
C;

Rank(E);
TorsionSubgroup(Jacobian(C));
```

Most of the time, the name of the function is the one you expect it to be \(with CamelCase\). One advantage of this approach is that lines of Magma code can sometimes be read like English sentences.


```magma
Factorization(Discriminant(MaximalOrder(F)));
```

If you are not sure about the name of a function, the [Magma handbook](https://magma.maths.usyd.edu.au/magma/handbook/) is the place to find more information. 

Basic sequences, like the integers between two given integers, are easy to define.


```magma
l := [1..10];
for i in l do 
    print i;
end for;
```

We can also define more complicated sequences compactly. 


```magma
pn := [ n : n in [1..100] | IsPrime(n^2-n+1) ];
pn;
[n^2-n+1 : n in pn];
```

Fluency with Magma \(or any computational algebra system!\) is a very useful skill to have! Being able to run up a piece of code to quickly test a conjecture on some range of data helps endlessly with research. Suppose you are Gauss and you are trying to derive the law of quadratic reciprocity. It is quite easy to check with a few lines. 


```magma
PP := Exclude(PrimesUpTo(1000),2);

for p in PP do 
    for q in PP do 
        if p ne q then 
            assert LegendreSymbol(p,q) * LegendreSymbol(q,p) eq (-1)^((p-1)/2 * (q-1)/2);
        end if;
    end for;
end for;
```

Magma has experts constantly refining and improving algorithms. Many quantities can be computed quickly in Magma, and there are lots of algorithms that can be improved by assumptions \(e.g. the general Riemann hypothesis\).


```magma
F1 := CyclotomicField(2^5);
time ClassNumber(F1);

delete F1;

SetClassGroupBounds("GRH");
F2 := CyclotomicField(2^5);
time ClassNumber(F2);
```
